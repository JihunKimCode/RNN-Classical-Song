```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('CSV/time.csv')
```

```python
# Create separate dataframes for CPU and GPU times
cpu_df = df[df['Type'] == 'CPU']
gpu_df = df[df['Type'] == 'GPU']

# Group by epoch and calculate the mean time for CPU and GPU dataframes
cpu_mean_times = cpu_df.groupby('Epochs').mean()['Time']
gpu_mean_times = gpu_df.groupby('Epochs').mean()['Time']

# Create line graphs of epoch vs. mean time for CPU and GPU
fig, ax = plt.subplots()
cpu_mean_times.plot(ax=ax, label='CPU', color = (114/255, 200/255, 189/255))
gpu_mean_times.plot(ax=ax, label='GPU', color = 'orange')

ax.set_xlabel('Epochs')
ax.set_ylabel('Time (minutes)')
ax.set_title('Epoch vs Mean Time in CPU and GPU')
ax.legend()
plt.show()
```

```python
# read the CSV file into a pandas DataFrame
df = pd.read_csv('CSV/time_sum.csv')

# create separate DataFrames for CPU and GPU data
df_cpu = df[df['Type'] == 'CPU']
df_gpu = df[df['Type'] == 'GPU']

# plot Total MIDI Play Time vs Time for CPU and GPU data
fig, ax = plt.subplots()
ax.plot(df_cpu['Total MIDI Play Time'], df_cpu['Time'], label='CPU', color = (114/255, 200/255, 189/255))
ax.plot(df_gpu['Total MIDI Play Time'], df_gpu['Time'], label='GPU', color = 'orange')
ax.set_xlabel('Total MIDI Playtime (minute)')
ax.set_ylabel('Running Time (second)')
ax.set_title('Total MIDI Playtime vs Time by CPU and GPU')
ax.legend()
plt.show()
```
