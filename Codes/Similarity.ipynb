```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CSV/MIDI_Similarity.csv')
```

```python
# filter out rows with missing similarity values
df = df[df['Similarity'] != 0]
```

```python
fig, ax = plt.subplots(figsize = (9,6))
df.boxplot(column='Similarity', by='Epoch', ax=ax, widths=0.2,
           color=dict(boxes='black', whiskers='black', medians='orange', caps='black'))

means = df.groupby('Epoch')['Similarity'].mean()
plt.plot(np.arange(len(means)) + 1, means, color = (114/255, 200/255, 189/255), marker='o', alpha = 1.0)

# Set x and y labels
plt.xlabel('Epoch')
plt.ylabel('Similarity')
plt.title("")
fig.suptitle('')
# fig.set_facecolor((0.87, 0.84, 0.77, 0.4))

ax.grid(visible=False)
# ax.set_facecolor((0.87, 0.84, 0.77, 0.4))

# Show the plot
plt.show()
```

```python
# Define the y = 0.4 line
y_threshold = 0.4
plt.axhline(y=y_threshold, color='gray', linestyle='--')

for composer in df['Name'].unique():
    composer_data = df[df['Name'] == composer]
    plt.plot(composer_data['Epoch'], composer_data['Similarity'], label=composer, marker = 'o')
    
# Add labels and legend to the plot
plt.title('Similarity by Epoch')
plt.xlabel('Epoch')
plt.ylabel('Similarity')
plt.legend(fontsize='small', loc = 'lower right')

# Show the plot
plt.show()
```
